(–ñ–∏–≤—ã–µ —É—á—ë–Ω—ã–µ –≤ –∫–Ω–∏–≥—É –Ω–µ –≤–∫–ª—é—á–µ–Ω—ã).

# –û—Ç–≤–µ—Ç—ã

## –†–µ—à–µ–Ω–∏–µ –∑–∞–¥–∞—á–∏ –æ –¥–≤–æ–π–Ω—ã—Ö —Ñ–∞–º–∏–ª–∏—è—Ö

**–û—Ç–≤–µ—Ç**: 16.

**–†–µ—à–µ–Ω–∏–µ**. –û–±–æ–∑–Ω–∞—á–∏–º ¬´–ø—Ä–æ—Å—Ç—ã–µ¬ª —Ñ–∞–º–∏–ª–∏–∏ –≤–Ω—É—Ç—Ä–∏ —Å–æ—Å—Ç–∞–≤–Ω—ã—Ö –±—É–∫–≤–∞–º–∏: `A, B, C, D`.

–≠—Ç–∏ —Ñ–∞–º–∏–ª–∏–∏ —É–∂–µ –º–æ–≥—É—Ç —è–≤–ª—è—Ç—å—Å—è –∏—Ö –æ–±—â–µ–π —Ñ–∞–º–∏–ª–∏–µ–π. –û—Å—Ç–∞–ª—å–Ω—ã–µ —Ñ–∞–º–∏–ª–∏–∏ ‚Äì [—Ä–∞–∑–º–µ—â–µ–Ω–∏—è](https://ru.wikipedia.org/wiki/%D0%9F%D0%B5%D1%80%D0%B5%D1%81%D1%82%D0%B0%D0%BD%D0%BE%D0%B2%D0%BA%D0%B0) –∏–∑ —á–µ—Ç—ã—Ä–µ—Ö –ø–æ –¥–≤—É–º. –ß–∏—Å–ª–æ —Ä–∞–∑–º–µ—â–µ–Ω–∏–π —Ä–∞–≤–Ω–æ $4!/(4-2)! = 4!/2! = 4\cdot3 = 12$. –î–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ, —ç—Ç–∏ 12 –≤–∞—Ä–∏–∞–Ω—Ç–æ–≤ –º–æ–∂–Ω–æ –∏ –ø–µ—Ä–µ–±—Ä–∞—Ç—å: 

`
AB AC AD
BA BC BD
CA CB CD
DA DB DC
`

–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, —Å—É–º–º–∞—Ä–Ω–æ–µ —á–∏—Å–ª–æ —Ñ–∞–º–∏–ª–∏–π —Ä–∞–≤–Ω–æ 16.

**–ü—Ä–æ–≥—Ä–∞–º–º–Ω–æ–µ —Ä–µ—à–µ–Ω–∏–µ** üåüüêç. –†–µ—à–∏–º –µ—â–µ –∏ –ø—Ä–æ–≥—Ä–∞–º–º–Ω–æ ‚Äì –Ω–∞ Python. –ú—ã –∫–∞–∫ —Ä–∞–∑ –Ω–µ–¥–∞–≤–Ω–æ —Ä–∞—Å—Å–º–∞—Ç—Ä–∏–≤–∞–ª–∏ –ø—Ä–∏–º–µ–Ω–µ–Ω–∏–µ [–±–∏–±–ª–∏–æ—Ç–µ–∫–∏ itertools](https://proglib.io/p/iteriruemsya-pravilno-20-priemov-ispolzovaniya-v-python-modulya-itertools-2020-01-03). –ü–æ–ø—Ä–∞–∫—Ç–∏–∫—É–µ–º—Å—è –≤ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–∏:

```python
from itertools import chain, permutations


def find_all_surnames(surnames):
    # —Ä–∞–∑–±–∏–≤–∞–µ–º —Å–æ—Å—Ç–∞–≤–Ω—ã–µ —Ñ–∞–º–∏–ª–∏–∏ –ø–æ –¥–µ—Ñ–∏—Å—É
    surnames = [surname.split('-') for surname in surnames]
    
    # –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –≤ –ø–ª–æ—Å–∫–∏–π —Å–ø–∏—Å–æ–∫, –ø–æ–ª—É—á–∞–µ–º —Å–ø–∏—Å–æ–∫ "–æ–±—ã—á–Ω—ã—Ö" —Ñ–∞–º–∏–ª–∏–π
    surnames = list(chain.from_iterable(surnames))
    
    # –æ—Å—Ç–∞–ª—å–Ω—ã–µ –≤–∞—Ä–∏–∞–Ω—Ç—ã —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—Ç –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∞–º –æ–±—ã—á–Ω—ã—Ö —Ñ–∞–º–∏–ª–∏–π
    perm_surnames = list(permutations(surnames, 2))
    
    # –æ–±—ä–µ–¥–∏–Ω—è–µ–º —Å–æ—Å—Ç–∞–≤–Ω—ã–µ —Ñ–∞–º–∏–ª–∏–∏ —á–µ—Ä–µ–∑ –¥–µ—Ñ–∏—Å
    perm_surnames = ['{}-{}'.format(i, j) for (i, j) in perm_surnames]
    
    # —Å–æ–µ–¥–∏–Ω—è–µ–º –¥–≤–∞ —Å–ø–∏—Å–∫–∞
    surnames.extend(perm_surnames)
    
    return surnames


# –ø—Ä–æ–≤–µ—Ä–∏–º —Ä–∞–±–æ—Ç—É —Ñ—É–Ω–∫—Ü–∏–∏ –Ω–∞ –ø—Ä–∞–∫—Ç–∏–∫–µ
surnames = find_all_surnames(['–°–∫–ª–æ–¥–æ–≤—Å–∫–∞—è-–ö—é—Ä–∏', '–ú–∏–∫–ª—É—Ö–æ-–ú–∞–∫–ª–∞–π'])

print('–û–±—â–µ–µ —á–∏—Å–ª–æ —Ñ–∞–º–∏–ª–∏–π —Ä–∞–≤–Ω–æ {}. –í–æ—Ç –∏—Ö —Å–ø–∏—Å–æ–∫ –±–µ–∑ —É—á—ë—Ç–∞ –ø–æ–ª–∞:'.format(len(surnames)))
for surname in surnames:
    print('- '+surname)
```

**–ï—Å–ª–∏ –≤–∞–º –ø–æ–Ω—Ä–∞–≤–∏–ª–∞—Å—å –∑–∞–¥–∞—á–∞**. –≠—Ç–æ —Ä–µ—à–µ–Ω–∏–µ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç –¥–≤–µ —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω—ã–µ —Ñ—É–Ω–∫—Ü–∏–∏ –≤—Å—Ç—Ä–æ–µ–Ω–Ω–æ–π –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ `itertools`. –ß—Ç–æ–±—ã –ø–æ–ø—Ä–∞–∫—Ç–∏–∫–æ–≤–∞—Ç—å—Å—è –≤ –∞–ª–≥–æ—Ä–∏—Ç–º–∞—Ö, —Ä–µ—à–∏—Ç–µ —Ç—É –∂–µ –∑–∞–¥–∞—á—É, –Ω–µ –∑–∞–¥–µ–π—Å—Ç–≤—É—è –±–∏–±–ª–∏–æ—Ç–µ–∫.

**–û—Ç–≤–µ—Ç**: 598.

**–†–µ—à–µ–Ω–∏–µ**. –§–∞–∫—Ç–∏—á–µ—Å–∫–∏ —ç—Ç–æ –∫–æ–º–±–∏–Ω–∞—Ç–æ—Ä–Ω–∞—è –∑–∞–¥–∞—á–∞ –Ω–∞ [–ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏](https://ru.wikipedia.org/wiki/%D0%9F%D0%B5%D1%80%D0%B5%D1%81%D1%82%D0%B0%D0%BD%D0%BE%D0%B2%D0%BA%D0%B0). –û–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–ª–æ–≤, —Å–æ—Å—Ç–∞–≤–ª–µ–Ω–Ω–æ–µ –∏–∑ —à–µ—Å—Ç–∏ –±—É–∫–≤, —Ä–∞–≤–Ω–æ `6! = 6¬∑5¬∑4¬∑3¬∑2¬∑1 = 720`. –£–¥–æ–±–Ω–µ–µ —Ä–∞—Å—Å–º–∞—Ç—Ä–∏–≤–∞—Ç—å —Ç–µ ¬´—Å–ª–æ–≤–∞¬ª, –∫–æ—Ç–æ—Ä—ã–µ —Å–ª–µ–¥—É—é—Ç –ø–æ—Å–ª–µ —Å–ª–æ–≤–∞ `TURING`. –ü–æ—Ç–æ–º—É —á—Ç–æ –∏—Ö –º–µ–Ω—å—à–µ.

–¢–∞–∫–∏–µ —Å–ª–æ–≤–∞ –∏–º–µ—é—Ç —Å–Ω–∞—á–∞–ª–∞ –≤–∏–¥ `TURN‚àó‚àó`, –ø–æ—Ç–æ–º —Å–º–µ–Ω—è—é—Ç—Å—è —Å–ª–æ–≤–∞–º–∏ –≤–∏–¥–∞ `U‚àó‚àó‚àó‚àó‚àó`. –ó–≤—ë–∑–¥–æ—á–∫–∞ –æ–±–æ–∑–Ω–∞—á–∞–µ—Ç –æ–¥–Ω—É –∏–∑ —à–µ—Å—Ç–∏ –±—É–∫–≤, –∫–æ—Ç–æ—Ä—ã—Ö –µ—â–µ –Ω–µ—Ç –≤ —Å–ª–æ–≤–µ. –î–ª—è –ø–µ—Ä–≤–æ–π –∫–æ–º–±–∏–Ω–∞—Ü–∏–∏ –±—É–∫–≤ –æ—Å—Ç–∞–µ—Ç—Å—è –≤–∞—Ä–∏–∞–Ω—Ç–æ–≤: `2!`, –¥–ª—è –≤—Ç–æ—Ä–æ–π ‚Äì `5!`. –¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ ¬´—Å–ª–æ–≤¬ª, —Å–ª–µ–¥—É—é—â–∏—Ö –∑–∞ `TURING` —Ä–∞–≤–Ω–æ `5! + 2! = 120 + 2 = 122`.

–ù–æ–º–µ—Ä —Ñ–∞–º–∏–ª–∏–∏ —ç—Ç–æ —á–∏—Å–ª–æ –æ–±—â–∏—Ö —Å–ª–æ–≤ –º–∏–Ω—É—Å –Ω–æ–º–µ—Ä —Å–ª–µ–¥—É—é—â–∏—Ö –∑–∞ –Ω–µ–π. –¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –µ—Å–ª–∏ –Ω—É–º–µ—Ä–∞—Ü–∏—è –Ω–∞—á–∏–Ω–∞–µ—Ç—Å—è —Å –µ–¥–∏–Ω–∏—Ü—ã, –Ω–æ–º–µ—Ä —Ñ–∞–º–∏–ª–∏–∏ —Ä–∞–≤–µ–Ω `720 ‚Äì 122 = 598`.

**–ü—Ä–æ–≥—Ä–∞–º–º–Ω–æ–µ —Ä–µ—à–µ–Ω–∏–µ** üåüüêç. –ï—Å–ª–∏ –≤—ã –Ω–µ —Å–º–æ–≥–ª–∏ —Ä–µ—à–∏—Ç—å –∑–∞–¥–∞—á—É –∞–Ω–∞–ª–∏—Ç–∏—á–µ—Å–∫–∏, –º–æ–∂–Ω–æ –±—ã–ª–æ —Ä–µ—à–∏—Ç—å –∑–∞–¥–∞—á—É –ø—Ä–æ–≥—Ä–∞–º–º–Ω–æ. –ù–∏–∂–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–æ –ø—Ä–æ—Å—Ç–æ–µ —Ä–µ—à–µ–Ω–∏–µ –Ω–∞ Python —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º –≤—Å—Ç—Ä–æ–µ–Ω–Ω–æ–π –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ [itertools](https://proglib.io/p/iteriruemsya-pravilno-20-priemov-ispolzovaniya-v-python-modulya-itertools-2020-01-03):

```python
from itertools import permutations


def find_surname_position(surname: str):
    surname = surname.upper()  # –µ—Å–ª–∏ —Ñ–∞–º–∏–ª–∏—è –Ω–∞–ø–∏—Å–∞–Ω–∞ –Ω–µ –∑–∞–≥–ª–∞–≤–Ω—ã–º–∏ –±—É–∫–≤–∞–º–∏
    letters = list(surname)  # —Å–ø–∏—Å–æ–∫ –±—É–∫–≤, –∏–∑ –∫–æ—Ç–æ—Ä—ã—Ö —Å–æ—Å—Ç–æ–∏—Ç —Ñ–∞–º–∏–ª–∏—è
    letters_vars = permutations(sorted(letters), len(surname))  # –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏
    words_vars = [''.join(item) for item in letters_vars]  # "—Å–ª–æ–≤–∞" –∏–∑ –±—É–∫–≤
    position = words_vars.index(surname) + 1  # –Ω—É–º–µ—Ä–∞—Ü–∏—è –Ω–∞—á–∏–Ω–∞–µ—Ç—Å—è —Å –µ–¥–∏–Ω–∏—Ü—ã
    return position

input_surname = input('–í–≤–µ–¥–∏—Ç–µ —Ñ–∞–º–∏–ª–∏—é: ')

req = '–ù–æ–º–µ—Ä —Ñ–∞–º–∏–ª–∏–∏ –≤ —Å–ø–∏—Å–∫–µ ¬´—Å–ª–æ–≤¬ª, —Å–æ—Å—Ç–∞–≤–ª–µ–Ω–Ω–æ–º –∏–∑ –±—É–∫–≤ —Ñ–∞–º–∏–ª–∏–∏: {}.'
print(req.format(find_surname_position(input_surname)))
```
