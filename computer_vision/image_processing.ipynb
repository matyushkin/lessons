Загрузим функции для загрузки, просмотра и сохранения изображений.

```python
from skimage.io import imread, imshow, imsave
```

Загрузим изображение и сохраним в переменной.

```python
img = imread('images/tiger-color.png')
```

Посмотри на загруженное изображение

```python
imshow(img)
```

Узнаем размер изображения:

```python
img.shape
```

Посмотрим на конкретный пиксель (на носу тигра):

```python
img[383, 374]
```

В конце указан тип чисел, в которых хранится цвет.

Изменим цвет пикселя на желтый и сохраним изображение.

```python
img[383, 374] = [255, 255, 0]
```

```python
imsave('tiger-yellow-nose.png', img)
```
